```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        base_line, height = map(int, input('밑변, 높이를 입력해주세요.').split (','))
        area = base_line * height
        result = '평행 사변형 넓이는 : ' + str(area)
        return result

    def trape(self):
        base_line, upper_line, height = map(int, input('밑변, 윗변, 높이를 입력해주세요.').split (','))
        area = (base_line + upper_line) * height * 0.5
        result = '사다리꼴 넓이는 : ' + str(area)
        return result

a = Square()
```

```python
a.par()
```

```python
a.trape()
```

```python
class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price

    def menu_print(self):
        for i in range (len(self.menu)):
            print (i+1, self.menu[i], ':', self.price[i],'원')

    def menu_select(self):
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요."))
            if 1 <= n and n <= len(menu):
                self.price_sum = self.price[n-1]
            else :
                print("없는 메뉴입니다.")
        t = 0
        while t != 1 and t != 2:
            t = int (input("Hot은 1, Ice는 2를 입력하세요."))
        if t == 1:
            pass
        elif t == 2:
            pass
        else:
            print("1과 2 중 하나 입력하세요.")

        self.order_menu = []
        self.order_price = []
        self.order_menu.append(temp[t-1] + '' + self.menu[n-1])
        print(temp[t-1], self.menu[n-1], ':', self.price_sum, '원')

        while n != 0:
            n = int(input ("추가 주문은 음료 번호를 누르시고, 지불하려면 0을 누르세요."))
        if n > 0 and n < len(self.menu) +1:
            self.order_price.append(self.price[n-1])
            self.price_sum = self.price[n-1]
            print('추가 주문 음료:', temp[t-1], self.menu[n-1], ':' , self.price_sum, '원', '합계', self.price_sum)
        else :
            if n == 0 :
                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price)
            else :
                print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        pay = ['현금', '카드']
        payment = input ('현금 / 카드 중 고르세요')
        if payment == '카드':
            print('IC칩 방향에 맞게 카드를 꽂아주세요')
        elif payment == '현금':
            print('직원을 호출하겠습니다.')
        else:
            print('다시 고르세요.')

    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        for i in range(len(temp)):
                print(temp[i], ' : ', price[i])
                print('합계 금액 :', sum(price))
        for i in range(5):
            print('|' + ' ' * 31+ '|')
            print('⟝' + '-' * 30 + '⟞')
```
