```python
##입출금 내역##

import random

class Account:
    total_accounts = 0

    def __init__(self, name, money):
        self.name = name
        self.money = money
        self.bank = "SC 은행"
        self.deposit_count = 0
        Account.total_accounts += 1
        self.bank_account = self.generate_account()
        self.transaction_history = []

    def generate_account(self):
        numbers1 = []
        for i in range (1,4):
            number = random.randint(1,9)
            numbers1.append(number)

        numbers2 = []
        for i in range (1,3):
            number = random.randint(1,9)
            numbers2.append(number)

        numbers3 = []
        for i in range (1,7):
            number = random.randint(1,9)
            numbers3.append(number)

        str_numbers1 = ''.join(map(str, numbers1))
        str_numbers2 = ''.join(map(str, numbers2))
        str_numbers3 = ''.join(map(str, numbers3))

        return f"{str_numbers1}-{str_numbers2}-{str_numbers3}"

    def deposit(self, plus_money):
        if plus_money > 0:
            self.money += plus_money
            self.deposit_count += 1
            self.transaction_history.append(f"입금: +{plus_money:,}, 잔고: {self.money:,}원")
            print(f"{self.name}님의 계좌에 {plus_money}원이 입금되었습니다.")

            if self.deposit_count >= 5:
                bonus_money = int(self.money * 0.01)
                self.money += bonus_money
                self.transaction_history.append(f"이자지급: {bonus_money:,}원, 잔고 :{self.money:,}원")
                print(f"입금 횟수가 {self.deposit_count}회가 되어 {bonus_money}원이 추가되었습니다.")
        else:
            print("입금 금액은 0보다 커야 합니다.")

    def withdraw(self, minus_money):
        if minus_money > 0 and minus_money <= self.money:
            self.money -= minus_money
            print(f"{self.name}님의 계좌에서 {minus_money}원이 빠져나갔습니다.")
            self.transaction_history.append(f"출금: -{minus_money:,} 잔고: {self.money:,}원")
        elif minus_money > self.money:
            print("출금 실패! 계좌에 충분한 돈이 없습니다.")
        else:
            print("출금 금액은 0보다 커야합니다.")

    def deposit_history(self):
        print(f"******{self.name}님의 입출금 내역******")
        if self.transaction_history:
            for transaction_history in self.transaction_history:
                print(transaction_history)
        else:
            print("입출금 내역이 없습니다.")

    @classmethod
    def get_total_accounts(cls):
        return cls.total_accounts

    def display_info(self):
        print("은행이름: ", self.bank)
        print("예금주: ", self.name)
        print("계좌번호: ",self.bank_account)
        print("잔고: ", f"{self.money:,}")
        print("입금 횟수: ", self.deposit_count, "회")


account1 = Account("이재훈", 100000)

account1.deposit(50000)  # 1회 입금
account1.deposit(100000)  # 2회 입금
account1.withdraw(30000)  # 출금
account1.deposit(200000)  # 3회 입금
account1.deposit(500000)  # 4회 입금
account1.deposit(1000000)  # 5회 입금 (이자 지급 발생)
account1.withdraw(100000)

account1.deposit_history()
```

```python
##attack_target 메서드 만들기##

import random

class Character:
    def __init__(self,name,health,defense,strike):
        self.name = name
        self.health = health
        self.defense = defense
        self.strike = strike


    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        real_damage = max(0, damage - self.defense)

        if real_damage > 0:
            self.health -= real_damage
            print(f"{self.name}이(가) {damage}의 공격을 받았지만 방어력 {self.defense} 덕분에 {real_damage}의 피해만 입었습니다. (남은 체력: {self.health})")
        else:
            print(f"{self.name}이(가) {damage}의 공격을 받았지만 방어력 {self.defense} 덕분에 피해를 입지 않았습니다. ")

        if not self.is_alive():
            print(f"{self.name}이(가) 쓰러졌습니다!")

    def attack_target(self,target):
        damage = random.randint(1, self.strike)
        print(f"{self.name}이(가) {target.name}을(를) {damage}공격합니다!")
        target.take_damage(damage)



#player = Character("대학원생", 50, 10, 40)
#monster = Character("교수", 40, 30, 20)
#player.attack_target(monster)
```

```python
##Player class##

class Player(Character):

    def __init__(self,name):
        super().__init__(name, 100, 5, 25)
        self.level = 1
        self.exp = 0

    def get_experience(self, exp):
        self.exp += exp
        print(f"{self.name}이(가) {exp} 경험치를 얻었습니다!")

        if exp >= 50:
            self.level += 1
            self.strike += 10
            self.defense += 5
            print(f"{self.name}이(가) 레벨업!(Lv.{self.level})  공격력이 10, 방어력이 5 상승했습니다.")

#player = Player("대학원생")

#player.attack_target(monster)
#player.get_experience(50)
```

```python
##Monster class##

class Monster(Character):
    def __init__(self, name, level):
        self.level = level

        health = random.randint(10,30) * level
        strike = random.randint(5,20) * level
        defense = random.randint(1,5) * level

        super().__init__(name, health, strike, defense)
```

```python
##battle함수##

def battle (player,monster):
    print(f"{player.name} VS {monster.name}")
    print("전투 시작")

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        print("졸업!")
        exp_get = monster.level * 20
        player.get_experience(exp_get)
    else:
        print("자네는 졸업 못하네!")
```

```python
##main 함수##
def main():
    monster_dict = {'부교수':1, '조교수':2, '교수':3}
    player_name = input('플레이어 이름을 입력하세요:')
    player = Player(player_name)

    for monster_name, level in monster_dict.items():
        print(f"{monster_name}이(가) 나타났다!")
        monster = Monster(monster_name, level)

        battle(player, monster)

        if not player.is_alive():
            print("자네는 졸업 못하네! [게임 오버]")
            return

    print("모든 교수님을 처치했습니다! 졸업 성공!")

if __name__ == "__main__":
    main()
```
